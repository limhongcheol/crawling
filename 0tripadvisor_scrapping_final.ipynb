```python
#진짜 코드
from tqdm import tqdm
from seleniumwire import webdriver
from time import sleep
import requests
import json
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException

def google_reviews(link_list,counts=10):
    revieww=[]
    count=0
    flag_textcase=0
    error_count=0
    try:
        service = Service(executable_path=r"C:\Users\Win10\Desktop\chromedriver-win64\chromedriver.exe")
        options = webdriver.ChromeOptions()
        #options.add_argument('--headless')
        options.add_argument('--no-sandbox')
        options.add_argument('--disable-dev-shm-usage')
        options.add_argument("user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36")
        driver = webdriver.Chrome(service=service, options=options)
        sleep(3)
        for store in link_list:
            driver.get('https://www.tripadvisor.cn/Attraction_Review-g297888-d320364-Reviews-Bulguksa_Temple-Gyeongju_Gyeongsangbuk_do.html')
            sleep(3)

            input_xpath = '//*[@id="__next"]/div/header/div/nav/div[1]/div/div[2]/form/input[1]'
            input_field = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, input_xpath)))
            input_field.send_keys(store+'\n')
            sleep(1)
            more_btn=driver.find_element("xpath", '//*[@id="search-filters"]/ul/li[4]/a')
            more_btn.click()

            sleep(1)
            more_btn2=driver.find_element("xpath",'//*[@id="__next"]/div/div[4]/div/div/div/div/div/div/div[1]/div/div/div/div[2]/div/div/div/div[2]/div/div/div[1]/div/div/div/div[2]/div/div[1]/span')
            more_btn2.click()

            driver.close()
            driver.switch_to.window(driver.window_handles[0])

            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            sleep(2)
            count_each=0
            revieww.append(store)
            flag_testcase=0
            flag_testcase2=0
            for j in range(10):
                for i in range(10):
                    try:
                        read_text = driver.find_element("id", f'comments_{i}')
                        revieww.append(read_text.text)
                        count+=1
                    except NoSuchElementException:
                        flag_testcase=1
                        break
                if(flag_testcase==1):
                    for i in range(1, 21):
                        try:
                            read_text = driver.find_element("xpath",f'//*[@id="component_10"]/div/div[3]/div[{i}]/div[3]/div[3]/div[1]/div[1]/q/span')
                            if read_text:
                                revieww.append(read_text.text)
                                flag_testcase2+=1
                                count+=1
                        except NoSuchElementException:
                            error_count+=1
                flag_testcase=0
                sleep(1)
                if(flag_testcase2==0):
                    try:
                        next_btn = driver.find_element("xpath", '//*[@id="tab-data-qa-reviews-0"]/div/div[5]/ul/li[6]')
                        next_btn.click()
                    except NoSuchElementException:
                        try:
                            next_btn = driver.find_element("xpath", '//*[@id="tab-data-qa-reviews-0"]/div/div[5]/ul/li[5]')
                            next_btn.click()
                        except NoSuchElementException:
                            next_btn = driver.find_element("xpath", '//*[@id="tab-data-qa-reviews-0"]/div/div[5]/ul/li[4]')
                        finally:
                            break
                if(flag_testcase2>0):
                    try:
                        next_btn=driver.find_element("xpath",'//*[@id="component_10"]/div/div[3]/ul/li[9]')
                        next_btn.click()
                    except ElementClickInterceptedException :
                        break
                sleep(1)
                revieww.append(count_each)
            revieww.append(count)
            print(store,count)
    except:
        print("ERROR")
        with open("reviews.txt", "w", encoding="utf-8") as f:
            for review in revieww:
                f.write(f"{review}\n")
    print("finish")
```
